```python
import pandas as pd
pd.set_option('display.max_columns', None) # Want to see all the output columns :)
```

```python
# Read the data
green_taxi_trips = pd.read_parquet("./green_tripdata_2025-11.parquet", engine = "pyarrow")
zone_lkup = pd.read_csv("./taxi_zone_lookup.csv")

display(green_taxi_trips.head(10), zone_lkup.head(10))
```

```python
# Merge the data sets for easier reference
trips_and_zones = green_taxi_trips.merge(
    zone_lkup,
    left_on = "PULocationID",
    right_on = "LocationID",
    how = "left")

trips_and_zones.head(10)
```

Question 3. Counting short trips

```python
november_trips = green_taxi_trips[
    (green_taxi_trips['lpep_pickup_datetime'] >= '2025-11-01') &
    (green_taxi_trips['lpep_pickup_datetime'] < '2025-12-01') &
    (green_taxi_trips['trip_distance'] <= 1)
]

short_trips_count = len(november_trips)
short_trips_count
```

Question 4. Longest trip for each day

```python
# Extract the date from lpep_pickup_datetime and find the longest trip for each day
longest_trip_per_day = green_taxi_trips[green_taxi_trips['trip_distance'] < 100].groupby(
    green_taxi_trips['lpep_pickup_datetime'].dt.date
)['trip_distance'].max()

# Find the day with the longest trip
day_with_longest_trip = longest_trip_per_day.idxmax()
longest_distance = longest_trip_per_day.max()

print(f"Day with longest trip: {day_with_longest_trip}")
print(f"Longest trip distance: {longest_distance} miles")
```

Question 5. Biggest pickup zone

```python
# Filter trips for November 18th, 2025
nov_18_trips = trips_and_zones[
    trips_and_zones['lpep_pickup_datetime'].dt.date == pd.to_datetime('2025-11-18').date()
]

# Group by Zone and sum the total_amount
pickup_zone_totals = nov_18_trips.groupby('Zone')['total_amount'].sum()

# Find the pickup zone with the largest total_amount
biggest_pickup_zone = pickup_zone_totals.idxmax()
largest_total = pickup_zone_totals.max()

print(f"Pickup zone with largest total_amount on Nov 18: {biggest_pickup_zone}")
print(f"Total amount: ${largest_total:.2f}")
```

Question 6. Largest tip

```python
# Filter trips with pickup zone "East Harlem North" in November 2025
east_harlem_north_pickups = trips_and_zones[
    (trips_and_zones['Zone'] == 'East Harlem North') &
    (trips_and_zones['lpep_pickup_datetime'].dt.month == 11) &
    (trips_and_zones['lpep_pickup_datetime'].dt.year == 2025)
]

east_harlem_north_pickups = trips_and_zones[trips_and_zones['trip_distance'] < 100].groupby(
    trips_and_zones['lpep_pickup_datetime'].dt.date
)['trip_distance'].max()


# Find the trip with the largest tip
max_tip_idx = east_harlem_north_pickups['tip_amount'].idxmax()
largest_tip_trip = east_harlem_north_pickups.loc[max_tip_idx]

dropoff_zone = largest_tip_trip['Zone_dropoff']
tip_amount = largest_tip_trip['tip_amount']

print(f"Dropoff zone with largest tip: {dropoff_zone}")
print(f"Tip amount: ${tip_amount:.2f}")
```
