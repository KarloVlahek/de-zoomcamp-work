```python
green_taxi_trips = pd.read_parquet("./green_tripdata_2025-11.parquet", engine = "pyarrow")
green_taxi_trips.head(10)
```

Question 3. Counting short trips

```python
november_trips = green_taxi_trips[
    (green_taxi_trips['lpep_pickup_datetime'] >= '2025-11-01') &
    (green_taxi_trips['lpep_pickup_datetime'] < '2025-12-01') &
    (green_taxi_trips['trip_distance'] <= 1)
]

short_trips_count = len(november_trips)
short_trips_count
```
